```python
data = requests.get('https://datausa.io/api/data?drilldowns=Nation&measures=Population').json()
```

```python
print(data)
```

```python
for item in data:
    print(item)
```

```python
information = data["data"]
print(information)
```

```python
for item in information:
    print(item)
```

```python
Populationtotal = sum(item["Population"] for item in information) 
print(Populationtotal)
```

```python
Yeartotal = avg(item["Year"] for item in information) 
print(Yeartotal)
```

```python
from statistics import mean
```

```python
Yearaverage = mean(item["Year"] for item in information) 
print(Yearaverage)
```

```python
years = [item["Year"] for item in information]
Yearaverage = mean(years)
print(Yearaverage)
```

```python
years = [int(item["Year"]) for item in information]
Yearaverage = mean(years)
print(Yearaverage)
```
